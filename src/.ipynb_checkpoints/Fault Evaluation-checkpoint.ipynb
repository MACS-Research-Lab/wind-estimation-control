```python
import numpy as np
from scipy.stats import norm
from scipy import stats

data = np.array([40, 63, 18, 54, 28, 27, 37, 42, 40, 21])  # Change this to your actual data

# Estimate mean and standard deviation
mean_estimation = np.mean(data)
std_estimation = np.std(data, ddof=1)  # ddof=1 for sample standard deviation

# Confidence interval for the mean
confidence_level = 0.95
z_score = stats.norm.ppf((1 + confidence_level) / 2)
margin_of_error_mean = z_score * (std_estimation / np.sqrt(len(data)))
confidence_interval_mean = (mean_estimation - margin_of_error_mean, mean_estimation + margin_of_error_mean)

# Confidence interval for the standard deviation
df = len(data) - 1  # degrees of freedom for sample standard deviation
alpha = 1 - confidence_level
chi2_lower = stats.chi2.ppf(alpha / 2, df)
chi2_upper = stats.chi2.ppf(1 - alpha / 2, df)
margin_of_error_std = np.sqrt((df * std_estimation**2) / chi2_lower), np.sqrt((df * std_estimation**2) / chi2_upper)
confidence_interval_std = (std_estimation - margin_of_error_std[0], std_estimation + margin_of_error_std[1])

# Print results
print("Mean Estimation:", mean_estimation)
print("Standard Deviation Estimation:", std_estimation)
print("Confidence Interval for Mean ({}%):".format(confidence_level * 100), confidence_interval_mean)
print("Confidence Interval for Standard Deviation ({}%):".format(confidence_level * 100), confidence_interval_std)
```

```python
square_np = np.array([[100,0,30], [100,100,30], [0,100,30], [0,0,30]])
```

```python
done = False
env = env_selector.get_env("lstm", lstm_params, [(0,0),(0,0),(0,0)], square_np, has_turbulence=False, start_alt=30)
state = env.reset()

log = DataLog(env.base_env.vehicle, env.base_env.ctrl,
                  other_vars=('reward',))

i = 0
vels = []
while not done:
    if i == 20:
        faulter = FaultInjector(env)
        env = faulter.inject_full_loss(0)
        
    action = lstm_agent.predict(state, deterministic=True)[0]
    vels.append(action)
    state, reward, done, info = env.step(action)
    print(info)
    log.log(reward=reward)
    i += 1

log.done_logging()
```

```python
plt.plot(log.x, log.y)
plt.xlabel("X (m)")
plt.ylabel("Y (m)")
plt.ylim(-5,20)
plt.title("Saturation Fault (3800 RPM) using RL Wind Controller")
```
